```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))
```

```python
scores = pd.read_csv('../../result/caseolap.csv')
cols = list(scores.columns)
entities = cols[0]
scores
```

```python
top10_scores = pd.DataFrame(columns = cols)
for i in range(0,10):
    top10_scores = top10_scores.append(pd.Series(0.0, index=cols), ignore_index=True)
top10_scores.columns = cols
top10_scores = top10_scores.drop(['protein'], axis=1)       #<---switch out with the entity column name depending on your entity
top10_entities = top10_scores.copy()

for i in range(1,len(cols)):
    categ_scores = list(scores[cols[i]])
    
    for j in range(0,10):
        TOP = max(categ_scores)
        TOP_ind = categ_scores.index(TOP)
        top10_scores.iloc[j,i-1] = TOP
        top10_entities.iloc[j,i-1] = scores[entities][TOP_ind]
        categ_scores[TOP_ind] = -1
        
top10_scores.index = range(1,11)
top10_scores
```

```python
top10_scores.to_csv('Data/top10_scores.csv')
```
