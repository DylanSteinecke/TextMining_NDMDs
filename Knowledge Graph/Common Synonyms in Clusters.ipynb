```python
# Stores every hit for the protein synonyms, i.e. every time there is a hit within the text documents for that synonym, the synonym is stored.
synonym_hit_list = []
f = open('../caseolap/log/entitycount_log.txt')
for line in f:
    if "#hits" in line:
        l = line.split(' #hits')[0].split(': ')[1]
        synonym_hit_list.append(l)
synonym_hit_list
```

```python
protein_syns_hit = {}
f = open('../caseolap/log/entitycount_log.txt')
for line in f:
    if "#hits" in line:
        ID = line.split('entity_rep ')[1].split(':')[0]
        protein_syns_hit[ID] = []

f = open('../caseolap/log/entitycount_log.txt')
for line in f:
    if "#hits" in line:
        syns = line.split(' #hits')[0].split(': ')[1]
        ID = line.split('entity_rep ')[1].split(':')[0]
        protein_syns_hit[ID].append(syns)
        
PSH_d = {}
PSHP_d = {}

for ID,syns in protein_syns_hit.items():
    PSH_d[ID] = {}
    PSHP_d[ID] = {}
for ID,syns in protein_syns_hit.items():
    syn_set = set(syns)
    d = {}
    dp = {}
    for syn in syn_set:
        d[syn] = syns.count(syn)
        dp[syn] = 0
    PSH_d[ID] = d
    PSHP_d[ID] = dp
    
#Stores ID with its hit synonyms and the proportion of hits which are that syn
for ID,syns in PSH_d.items():
    hit_tot = 0
    for syn,hits in syns.items():
        hit_tot += hits
    for syn,hits in syns.items():
        PSHP_d[ID][syn] = PSH_d[ID][syn]/hit_tot
```
